# Encoding different objects into PyTorch Tensors

## 1. Images

```python
# load a jpeg into Python

import torch
from torchvision import transforms
from PIL import Image

img_raw = Image.open("../../../data/random_images/maine_coon/maine_coon.jpg")
width, height = img_raw.size
img_raw.resize((int(width/9), int(height/9)))
```

```python
#convert image into a Pytorch tensor

convert_tensor = transforms.ToTensor()
img_tensor = convert_tensor(img_raw)
print(type(img_tensor))
print(img_tensor.shape)
img_tensor
```

```python
#There is a easier way with torchvision

import torchvision

img_tensor_2 = torchvision.io.decode_image("../../../data/random_images/maine_coon/maine_coon.jpg")
print(type(img_tensor_2))
print(img_tensor_2.shape)
img_tensor_2
```

Note that `torchvision.io` can also do it with other visual media such as videos.

The difference between `torchvision.io.decode_image` vs `torchvision.transforms.ToTensor` is that `ToTensor` returns a FloatTensor after normalizing to the range [0,1]. As per the documentation, it "converts a PIL Image or numpy.ndarray (H x W x C) in the range [0, 255] to a torch.FloatTensor of shape (C x H x W) in the range [0.0, 1.0]". Meanwhile, `io.decode_image` doesn't standardize.
