# Singular Value Decomposition

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/movie_lens/rating.csv", dtype={"userId": str, "movieId": str})
data.head()
```

```python
user_item = data[
    (data["userId"].astype(int) < 200)
    # & (data["movieId"].astype(int) < 100)
].groupby(["userId", "movieId"])["rating"].first().unstack(fill_value=0.0) # fill null with 0s
user_item.head()
```

```python
from scipy.sparse.linalg import svds

print(user_item.shape)

user_f, features, f_movie = svds(user_item.to_numpy())
pd.DataFrame(user_f)
```

```python
pd.DataFrame(features)
pd.DataFrame(np.diag(features))
```
